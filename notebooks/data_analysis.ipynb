```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../../data/raw_data/raw.csv')
X=df.drop(columns=['silica_concentrate','date'], axis=1)
y=df['silica_concentrate']
```

```python
# Analyse visuelle
for col in X.columns:  
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=X[col], y=y)  
    plt.title(f"Relation entre {col} et la cible")
    plt.show()
```

```python
# Analyse de coeff de Parson
corr_matrix = pd.concat([X, y], axis=1).corr()
print(corr_matrix["silica_concentrate"]) 
```

Les scattersplots sont des nuages sans forme et les coefficient de correlation sont proche de 0.

A partir de ces resultats on peut conclure que la relation n'est pas entre les features et la variable cible n'est pas linéaire
